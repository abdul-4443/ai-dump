```python
import pandas as pd
import numpy as np

basket = pd.read_csv('customers.csv')

basket.isnull().sum()   
threshold = basket.shape[1] - 2
basket = basket.dropna(thresh=threshold)

basket['Age'] = basket['Age'].fillna(basket['Age'].median())

basket['City'] = basket['City'].fillna(basket['City'].mode()[0])

basket['Annual_Income'] = (
    basket['Annual_Income']
    .astype(str)
    .str.replace(',', '', regex=False)
)
basket['Annual_Income'] = pd.to_numeric(basket['Annual_Income'], errors='coerce')

basket = basket.drop_duplicates()

basket['Gender'] = basket['Gender'].map({'Male': 0, 'Female': 1})

basket.head()
```

```python
ward = pd.read_csv('patients.csv')

ward['Admission_Date'] = pd.to_datetime(ward['Admission_Date'], errors='coerce', dayfirst=True)
ward['Discharge_Date'] = pd.to_datetime(ward['Discharge_Date'], errors='coerce', dayfirst=True)

severity_ladder = {'Low': 1, 'Medium': 2, 'High': 3}
ward['Severity'] = ward['Severity'].map(severity_ladder)

ward = pd.get_dummies(ward, columns=['Department'])

ward.dtypes
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split

dwelling = pd.read_csv('housing.csv')

blueprint, appraisal = train_test_split(dwelling, test_size=0.2, random_state=42)

minmax_pipe = MinMaxScaler()
blueprint[['Area', 'Distance_from_City']] = minmax_pipe.fit_transform(blueprint[['Area', 'Distance_from_City']])
appraisal[['Area', 'Distance_from_City']] = minmax_pipe.transform(appraisal[['Area', 'Distance_from_City']])

zscore_pipe = StandardScaler()
blueprint[['Age_of_House']] = zscore_pipe.fit_transform(blueprint[['Age_of_House']])
appraisal[['Age_of_House']] = zscore_pipe.transform(appraisal[['Age_of_House']])

blueprint.head()
```

```python
dispatch = pd.read_csv('orders.csv')

print('Duplicate rows:', dispatch.duplicated().sum())
dispatch = dispatch.drop_duplicates()

dispatch['Delivery_Time'] = dispatch['Delivery_Time'].fillna(dispatch['Delivery_Time'].mean())
dispatch['Customer_Rating'] = dispatch['Customer_Rating'].fillna(dispatch['Customer_Rating'].median())

dispatch = pd.get_dummies(dispatch, columns=['Cuisine_Type'])

print(dispatch.isnull().sum())
dispatch.head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

roster = pd.read_csv('students.csv')

roster['Marks'] = roster['Marks'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
roster['Marks'] = roster['Marks'].fillna(roster['Marks'].mean())

roster = roster.drop_duplicates()

grade_ladder = {'D': 1, 'C': 2, 'B': 3, 'A': 4}
roster['Grade'] = roster['Grade'].map(grade_ladder)

verdict_enc = LabelEncoder()
roster['Passed'] = verdict_enc.fit_transform(roster['Passed'])

hour_scaler = StandardScaler()
roster[['Study_Hours']] = hour_scaler.fit_transform(roster[['Study_Hours']])

roster.head()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

ledger = pd.DataFrame({
    'Student_ID': [1, 2, 3, 4, 2],
    'Course': ['AI', 'Web Dev', 'AI', None, 'Web Dev'],
    'Completion_Percentage': ['80', '90%', None, '70', '90%'],
    'Enroll_Date': ['2024-01-10', '10/02/2024', 'invalid', '2024-03-01', '10/02/2024'],
    'Feedback': ['Good', 'Excellent', 'Average', 'Good', 'Excellent']
})

print(ledger.isnull().sum())

ledger = ledger.drop_duplicates()

ledger['Completion_Percentage'] = (
    ledger['Completion_Percentage']
    .astype(str)
    .str.replace('%', '', regex=False)
    .replace('None', np.nan)
)
ledger['Completion_Percentage'] = pd.to_numeric(ledger['Completion_Percentage'], errors='coerce')

ledger['Enroll_Date'] = pd.to_datetime(ledger['Enroll_Date'], errors='coerce', dayfirst=True)

ledger['Completion_Percentage'] = ledger['Completion_Percentage'].fillna(ledger['Completion_Percentage'].mean())
ledger['Course'] = ledger['Course'].fillna(ledger['Course'].mode()[0])

ledger = pd.get_dummies(ledger, columns=['Course'])

feedback_enc = LabelEncoder()
ledger['Feedback'] = feedback_enc.fit_transform(ledger['Feedback'])

pct_scaler = MinMaxScaler()
ledger[['Completion_Percentage']] = pct_scaler.fit_transform(ledger[['Completion_Percentage']])

ledger.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

shelf = pd.DataFrame({
    'Product_ID': [101, 102, 103, 104, 101],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Grocery', 'Electronics'],
    'Price': ['1000', '2000', 'invalid', '500', '1000'],
    'Quantity_Sold': [5, 10, None, 8, 5],
    'Discount': [10, 20, 15, None, 10],
    'Purchase_Date': ['2024/01/01', '01-02-2024', '2024-03-01', 'invalid', '2024/01/01']
})

shelf = shelf.drop_duplicates()

shelf['Price'] = pd.to_numeric(shelf['Price'], errors='coerce')

shelf['Purchase_Date'] = pd.to_datetime(shelf['Purchase_Date'], errors='coerce', dayfirst=True)

shelf['Quantity_Sold'] = shelf['Quantity_Sold'].fillna(shelf['Quantity_Sold'].median())
shelf['Discount'] = shelf['Discount'].fillna(shelf['Discount'].mean())

shelf = pd.get_dummies(shelf, columns=['Category'])

numeric_features = ['Price', 'Quantity_Sold']
rack_train, rack_test = train_test_split(shelf, test_size=0.2, random_state=42)

volume_scaler = StandardScaler()
rack_train[numeric_features] = volume_scaler.fit_transform(rack_train[numeric_features])
rack_test[numeric_features] = volume_scaler.transform(rack_test[numeric_features])

rack_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

vault = pd.read_csv('banking.csv')

print(vault.isnull().sum())

vault = vault.drop_duplicates()

vault['Income'] = (
    vault['Income']
    .astype(str)
    .str.replace(',', '', regex=False)
)
vault['Income'] = pd.to_numeric(vault['Income'], errors='coerce')

vault['Age'] = vault['Age'].astype(str).replace('unknown', np.nan)
vault['Age'] = pd.to_numeric(vault['Age'], errors='coerce')

vault['Age'] = vault['Age'].fillna(vault['Age'].median())
vault['Credit_Score'] = vault['Credit_Score'].fillna(vault['Credit_Score'].mean())
vault['City'] = vault['City'].fillna(vault['City'].mode()[0])

status_enc = LabelEncoder()
vault['Loan_Status'] = status_enc.fit_transform(vault['Loan_Status'])

vault = pd.get_dummies(vault, columns=['City'])

risk_scaler = StandardScaler()
vault[['Income', 'Credit_Score']] = risk_scaler.fit_transform(vault[['Income', 'Credit_Score']])

vault.head()
```
