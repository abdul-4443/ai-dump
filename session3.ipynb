```python
import pandas as pd
data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Temperature": [22, 20, 19, 21, 21, 25, 23],
    "Rainfall": [0.0, 5.2, 12.1, 1.5, 0.0, 0.0, 0.8]
}

df = pd.DataFrame(data)
print(f"Mean: {df["Temperature"].mean()}")
print(f"Median: {df["Temperature"].median()}")
print(f"Mode: {df["Temperature"].mode().iloc[0]}")
print(f"Range: {df["Temperature"].max() - df["Temperature"].min()}")

print("it tells us that how the weather was most of the time")
```
